3. Найти пределы

$$\lim_{x → 5}\frac{x^2-6x+5}{x^2-25} = \lim_{x → 5}\frac{(x-1)(x-5)}{(x-5)(x+5)} = \lim_{x → 5}\frac{x-1}{x+5} = \frac{4}{10}$$

$$\lim_{x → -1}\frac{x^3+x+2}{x^3+1} = \lim_{x → -1}\frac{(x+1)(x^2-x+2)}{(x+1)(x^2-x+1)} = \frac{1+1+2}{1+1+1} = \frac{4}{3}$$

$$\lim_{x → 3}\frac{\sqrt{2x+3}-3}{\sqrt{x-2}-1} = \lim_{x → 3}\frac{(\sqrt{2x+3}-3)(\sqrt{2x+3}+3)}{(\sqrt{x-2}-1)(\sqrt{x-2}+1)} = \lim_{x → 3}\frac{(2x-6)(\sqrt{x-2}+1)}{(x-3)(\sqrt{2x+3}+3)} = \lim_{x → 3}\frac{2(\sqrt{x-2}+1)}{\sqrt{2x+3}+3} = \frac{2 \cdot 2}{3 + 3} = \frac{2}{3}$$

$$\lim_{x → 0}\frac{1-cosx}{x^2} = \lim_{x → 0}\frac{2sin^2(\frac{x}{2})}{x^2} = 2\lim_{x → 0}\frac{\left(\frac{sin(\frac{x}{2})}{\frac{x}{2}}\right)^2}{4} = 2 \cdot \frac{1}{4} = \frac{1}{2}$$

$$\lim_{x → 0}(x \cdot ctg(x)) = \lim_{x → 0}\frac{cosx \cdot x}{sinx} = \lim_{x → 0}\frac{cosx}{\frac{sinx}{x}} = \lim_{x → 0}cosx = 1$$

$$\lim_{x → 0}\sqrt[2x]{1+3x} = \lim_{x → 0}(1+3x)^\frac{1}{2x} = \lim_{x → 0}(1+3x)^{\frac{1}{2x} \cdot \frac{3}{3}} = \lim_{x → 0}(1+3x)^{\frac{1}{3x} \cdot \frac{3}{2}} = e^\frac{3}{2} = e\sqrt{e}$$





$$\lim_{x → 0}\left(\frac{3+5x}{3+2x}\right)^\frac{1}{x} = \lim_{x → 0}\left(1 + \frac{3x}{3+2x}\right)^\frac{1}{x} = \lim_{x → 0}\left(1 + \frac{3x}{3+2x}\right)^{\frac{2x+3}{3x} \cdot \frac{3}{2x+3}} = \lim_{x → 0} e^{\frac{3}{2x+3}} = e^1 = e$$
