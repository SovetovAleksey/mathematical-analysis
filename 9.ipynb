1. Исследовать ряд на сходимость, используя признак д'Аламбера

$$\sum_{n=1}^{\infty} \frac{n^n}{(n!)^2}$$

Найдем предел

$$\lim_{n → \infty}\frac{\frac{(n+1)^{n+1}}{((n+1)!)^2}}{\frac{n^n}{(n!)^2}} = \lim_{n → \infty}\frac{(n+1)^{n+1} \cdot (n!)^2}{((n+1)!)^2 \cdot n^n} = \lim_{n → \infty}\frac{(n+1)^n(n+1)(n!)^2}{(n+1)! \cdot n!(n+1) \cdot n^n} = \lim_{n → \infty}\frac{(n+1)^n \cdot n!}{(n+1)! \cdot n^n} = \lim_{n → \infty}\frac{(n+1)^n}{(n+1) \cdot n^n} = \lim_{n → \infty}(n+1)^{n-1} \cdot \left(\frac{1}{n}\right)^n = 0 < 1 ⇒ $$ ряд сходится

2. Исследовать ряд на сходимость, используя радикальный признак Коши

$$\lim_{n → \infty}\sqrt[n]\frac{n}{2^n} = \lim_{n → \infty}\frac{\sqrt[3]n}{2} = \frac{1}{2}\lim_{n → \infty}\sqrt[n]{n} = \frac{1}{2}\lim_{n → \infty}n^{\frac{1}{n}} = \frac{1}{2} < 1 ⇒$$ ряд сходится

3. Исследовать ряд на сходимость, используя признак Лейбница
